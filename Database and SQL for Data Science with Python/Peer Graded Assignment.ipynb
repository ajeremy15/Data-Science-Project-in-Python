## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
%%sql problem_1 <<
SELECT COUNT(*) AS TOTAL_NUM_OF_CRIMES
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA
```

```python
problem_1
```

### Problem 2

##### List community areas with per capita income less than 11000.


```python
%%sql problem_2 <<
SELECT COMMUNITY_AREA_NAME
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA
WHERE PER_CAPITA_INCOME < 11000
```

```python
problem_2
```

### Problem 3

##### List all case numbers for crimes  involving minors?(children are not considered minors for the purposes of crime analysis)


```python
%%sql problem_3 <<
SELECT CASE_NUMBER
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA
WHERE DESCRIPTION NOT LIKE '%CHILD%' OR PRIMARY_TYPE NOT LIKE 'CHILDREN%'
```

```python
problem_3
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%%sql problem_4 <<
SELECT *
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA
WHERE PRIMARY_TYPE = 'Kidnapping' AND DESCRIPTION LIKE 'CHILD%'
```

```python
problem_4
```

### Problem 5

##### What kinds of crimes were recorded at schools?


```python
%%sql problem_5 <<
SELECT DISTINCT PRIMARY_TYPE
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA
WHERE LOCATION_DESCRIPTION LIKE 'SCHOOL%'
```

```python
problem_5
```

### Problem 6

##### List the average safety score for all types of schools.


```python
%%sql problem_6 <<
SELECT AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE, Elementary_Middle_or_High_School
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_PUBLIC_SCHOOLS
GROUP BY Elementary_Middle_or_High_School
```

```python
problem_6
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%%sql problem_7 <<
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
OFFSET 0 ROWS FETCH FIRST 5 ROWS ONLY
```

```python
problem_7
```

### Problem 8

##### Which community area is most crime prone?


```python
%%sql problem_8 <<
SELECT COMMUNITY_AREA_NUMBER
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA
WHERE COMMUNITY_AREA_NUMBER IN (SELECT TOP 1 COMMUNITY_AREA_NUMBER FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA 
                                WHERE COMMUNITY_AREA_NUMBER IS NOT NULL GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC)
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
%%sql problem_9 <<
SELECT COMMUNITY_AREA_NAME
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA
WHERE HARDSHIP_INDEX IN (SELECT MAX(HARDSHIP_INDEX) FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA)
```

```python
problem_9
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql problem_10 <<
SELECT COMMUNITY_AREA_NAME
FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CENSUS_DATA
WHERE COMMUNITY_AREA_NUMBER IN (SELECT TOP 1 COMMUNITY_AREA_NUMBER FROM IBMDataScienceCourseSQLDatabase.dbo.CHICAGO_CRIME_DATA 
                                WHERE COMMUNITY_AREA_NUMBER IS NOT NULL GROUP BY COMMUNITY_AREA_NUMBER ORDER BY COUNT(*) DESC)
```
